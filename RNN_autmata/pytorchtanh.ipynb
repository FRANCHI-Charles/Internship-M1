```python
Q = 3
SIGMA = 3
A = Q * SIGMA
size = (Q*SIGMA)**2


test = np.zeros((size, size))
for i in range(size):
    test[i, i%A * A: (i%A+1)*A] = [1 if i//A ==j else -1 for j in range(A)]


np.linalg.inv(test)
```
